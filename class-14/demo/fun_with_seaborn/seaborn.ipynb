```python
flights = sns.load_dataset("flights")
flights
```

```python
type(flights)
```

```python
sns.relplot(x="year", y="passengers", data=flights)
sns.relplot(x="year", y="passengers", data=flights, hue="month")
```

```python
year_sums = flights.groupby("year", as_index=False).sum()
```

```python
year_sums
```

```python
sns.relplot(x="year", y="passengers", data=year_sums)
```

```python
sns.lmplot(x="year", y="passengers", data=year_sums)
```

```python
sns.barplot(x="month", y="passengers", data=flights)
```
